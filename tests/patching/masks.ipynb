```python
import os
import json
os.environ["CUDA_VISIBLE_DEVICES"]=""
os.environ["XLA_PYTHON_CLIENT_PREALLOCATE"]="false"
import jax.numpy as jnp
from itertools import product
import matplotlib.pyplot as plt
```

```python
ldim = 5
patch_size = 3

localHilbertDim = ldim**patch_size
maxParticleP1 = (ldim-1)*patch_size +1
patch_states = jnp.array(list(product(range(ldim),repeat=patch_size)))

particles = jnp.sum(patch_states,axis=1)
particles
patch_states[3]
```

```python
must_mask = 2*jnp.array([particles<j for j in range(maxParticleP1)])
plt.figure(figsize=(15,6))
plt.imshow(must_mask)
plt.colorbar()
```

```python
must_mask_no = 2 * jnp.tril(jnp.ones((ldim,ldim)),k=-1)
plt.imshow(must_mask_no)
plt.colorbar()
```

```python
can_mask_no = jnp.flip(must_mask_no)

plt.imshow(can_mask_no)
plt.colorbar()
```
